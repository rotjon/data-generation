$$ \frac{\partial T}{\partial t} = \frac{1}{\rho C_p} \left[ \frac{\partial }{\partial x_n} \lambda \left( \frac{\partial T}{\partial x_n} \right) \right] $$

$$ \frac{ \rho C_P}{k} \frac{\partial T}{\partial t} = \frac{\partial ^2 T}{\partial x^2} $$

$T(x_{i+1})=T(x_i)+\dfrac{\delta x F(x_i)}{k}$

```python
import pandas as pd
import numpy as np
```

```python
def readDatafile(datafile):
    df1 = pd.read_csv(datafile + ".txt", skiprows= 8, sep=r'\s{2,}', engine='python', header=None )
    df1[0] = df1[0].round(3)
    df1.set_index(0, inplace = True)
    df1.index.name = 'Time'
    df1.columns = [datafile]
    daterng = pd.date_range(start = '2019-11-15', periods =df1.size,freq='50ms')
    df1['date'] = daterng
    df1.set_index('date', inplace=True)
    df1 = df1.resample('1ms').interpolate('cubic')
    return df1
```

```python
df1 = readDatafile("cooledVariable")
df2 = readDatafile("middle")
df3 = readDatafile("measureVariable")
df4 = readDatafile("1mmAbove")
df5 = readDatafile("1mmbelov")
```

```python
result = pd.concat([df1,df2,df3, df4, df5], axis=1, sort=False)
```
